```r
# --- Vectors and Operations ---
vec1 <- c(1, 2, 3, 4, 5)
vec2 <- c(10, 20, 30, 40, 50)
sum_vec <- vec1 + vec2
cat('Sum of vectors:', sum_vec, '\n')
mean(vec1)
max(vec2)
min(vec1)
vec3 <- c(1, 2)
vec4 <- c(10, 20, 30, 40)
vec3 + vec4
```

```r
# --- Logical Operations and Missing Values ---
x <- c(TRUE, TRUE, FALSE)
all(x)
any(x)
data <- c(1, 2, NA, 4, 5, NA)
is.na(data)
clean_data <- na.omit(data)
cat('Cleaned Data:', clean_data, '\n')
data[is.na(data)] <- mean(data, na.rm = TRUE)
cat('Replaced NA with mean:', data, '\n')
```

```r
# --- Filtering and Conditional ---
nums <- c(5, 10, 15, 20, 25, 30)
filtered <- nums[nums > 15]
cat('Numbers greater than 15:', filtered, '\n')
scores <- c(45, 67, 89, 32, 50)
result <- ifelse(scores >= 50, 'Pass', 'Fail')
cat('Results:', result, '\n')
```
